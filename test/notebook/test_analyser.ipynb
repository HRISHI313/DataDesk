```python
def total_record(dataframe):
    return len(dataframe)

total_record(df)
```

```python
def total_unique_retailers(dataframe):
    return dataframe[LIST_NAME_COL].unique()

total_unique_retailers(df)
```

```python
def total_unique_retailers_count(dataframe):
    return dataframe[LIST_NAME_COL].nunique()

total_unique_retailers_count(df)
```

```python
def records_per_retailer(dataframe):
    return dataframe[LIST_NAME_COL].value_counts()

records_per_retailer(df)
```

```python
def construction_flag_breakdown(dataframe):
    # fill NaN in construction flag with NORMAL label first
    df = dataframe.copy()
    df[CONSTRUCTION_FLAG_COL] = df[CONSTRUCTION_FLAG_COL].fillna(NORMAL)
    
    # group and count
    breakdown_df = df.groupby([LIST_NAME_COL, CONSTRUCTION_FLAG_COL]).size().unstack(fill_value=0)
    
    # add total column
    breakdown_df["Total"] = breakdown_df.sum(axis=1)
    
    return breakdown_df

construction_flag_breakdown(df)
```

```python
def polygon_status_breakdown(dataframe):
    # fill NaN in polygon status with NORMAL label first
    df = dataframe.copy()
    df[POLYGON_STATUS_COL] = df[POLYGON_STATUS_COL].fillna(NORMAL)
    
    # group and count
    polygon_status_df = df.groupby([LIST_NAME_COL, POLYGON_STATUS_COL]).size().unstack(fill_value=0)
    
    # add total column
    polygon_status_df["Total"] = polygon_status_df.sum(axis=1)
    
    return polygon_status_df

polygon_status_breakdown(df)
```

```python
def duplicate_ali_analysis(df):
    
    duplicates = df[df[ALI_COL].duplicated(keep=False)]
    
    duplicate_count = df[ALI_COL].duplicated().sum()
    unique_count    = df[ALI_COL].nunique()
    total           = len(df)
    
    return {
        "total"           : total,
        "unique_ali"      : unique_count,
        "duplicate_count" : duplicate_count,
        "duplicate_rows"  : duplicates    # actual duplicate records
    }

duplicate_ali_analysis(df)
```

```python
def get_full_analysis(df):
    return {
        "total_records"     : total_record(df),
        "unique_retailers"  : total_unique_retailers(df),
        "per_retailer"      : records_per_retailer(df),
        "construction_flag" : construction_flag_breakdown(df),
        "polygon_status"    : polygon_status_breakdown(df),
        "duplicate_ali"     : duplicate_ali_analysis(df)  # ← added
    }
```
